Importing Open CV

```python
import cv2
```

Importing Numpy

```python
import numpy as np
```

Importing Mediapipe and related styles from Mediapipe

```python
import mediapipe as mp
mp_drawing = mp.solutions.drawing_utils
mp_drawing_styles = mp.solutions.drawing_styles
mp_hands = mp.solutions.hands
```

Importing tensorflow and model from tensorflow

```python
import tensorflow as tf
from tensorflow.keras.models import load_model
```

```python
# initialize mediapipe
mpHands = mp.solutions.hands
hands = mpHands.Hands(max_num_hands=1, min_detection_confidence=0.7)
mpDraw = mp.solutions.drawing_utils
```

Loading the Hand Gesture Recognition model

```python
# Load the gesture recognizer model
model = load_model('mp_hand_gesture')

# Load class names
f = open('gesture.names', 'r')
classNames = f.read().split('\n')
f.close()
print(classNames)
```

```python
# Initialize the webcam for Hand Gesture Recognition Python project
cap = cv2.VideoCapture(0)

while True:
  # Read each frame from the webcam
  _, frame = cap.read()

    # Flip the frame vertically
  frame = cv2.flip(frame, 1)

  #Making the Green Box for printing threshold output
  cv2.rectangle(frame,(300,300),(100,100),(0,255,0),0)

  #apturing the frame in the Geen Box
  crop_img=frame[100:300,100:300]

  #Converting the colour of the part of the fram in the green box
  grey=cv2.cvtColor(crop_img,cv2.COLOR_BGR2GRAY)
  value = (35,35)

  #Blurring the Frames using GaussianBlurr to the part of the frame in the green box
  blurred = cv2.GaussianBlur(grey,value,0)

   #Applying OTSU's Thresholding method of Open CV
  _, thresh1 = cv2.threshold(blurred,127,255,cv2.THRESH_BINARY_INV+cv2.THRESH_OTSU)

  #Displaying the Threshold 
  cv2.imshow('Threshold',thresh1)


  x , y, c = frame.shape

  #converting the color
  framergb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
  
  # Get hand landmark prediction, using mediapipe
  result = hands.process(framergb)

  className = ''

  # post process the result
  if result.multi_hand_landmarks:
      landmarks = []
      for handslms in result.multi_hand_landmarks:
          for lm in handslms.landmark:
              # print(id, lm)
              lmx = int(lm.x * x)
              lmy = int(lm.y * y)

              landmarks.append([lmx, lmy])

          # Drawing landmarks on frames
          mpDraw.draw_landmarks(
            frame, 
            handslms,
            mpHands.HAND_CONNECTIONS,
            mp_drawing_styles.get_default_hand_landmarks_style(),
            mp_drawing_styles.get_default_hand_connections_style())

          # Predict gesture in Hand Gesture Recognition project
          prediction = model.predict([landmarks])
          # print(prediction)
          classID = np.argmax(prediction)
          className = classNames[classID]

          print(className)

  # show the prediction on the frame
  cv2.putText(frame, className, (10, 50), cv2.FONT_HERSHEY_SIMPLEX,
                 1, (0,0,255), 2, cv2.LINE_AA)

  # Show the final output
  cv2.imshow("Output", frame)

  #Press 'q' on keyboard to stop the hand recognition process
  if cv2.waitKey(1) == ord('q'):
        break

# release the webcam and destroy all active windows
cap.release()
cv2.destroyAllWindows()
```
